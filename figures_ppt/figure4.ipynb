```python
import matplotlib.pyplot as plt
import numpy as np
import time

label_font_conf = {
    # "weight": "bold",
    "size": "18"
}
labels = ['rust', 'bugs', 'build', 'drivers', 'abstraction']
# 
dates = ['23/7', '23/3', '23/1', '22/10', '22/7', '22/4', '22/1']
dates.reverse()
# 
percentages = np.array([
    [57.55814, 65.92357, 55.30864, 38.91213, 26.31579, 57.57576, 18.84058],
    [26.45349, 9.12951, 4.69136, 15.06276, 15.78947, 9.09091, 19.32367],
    [12.7907,7.431,33.08642,43.51464,57.89474,33.33333,58.9372],
    [0.5814, 1.06157, 3.20988, 0.83682, 0, 0, 2.89855],
    [2.61628, 16.45435, 3.7037, 1.67364, 0, 0, 0],
])

current_time = int(time.time())
np.random.seed(current_time)
process = np.random.rand(5, 7) * 100
# 
fig, axs = plt.subplots(1, 2, figsize=(9,5),dpi=100)

# 
markers = ['o', '^', 'x', '+', '.']
colors = ['b', 'r', 'g', 'y', 'purple']
for i, per in enumerate(percentages):
    per = per[::-1]
    axs[0].plot(dates, per, marker=markers[i], linestyle='-', color=colors[i], linewidth=1.2, label=labels[i])

for i, pro in enumerate(process):
    axs[1].plot(dates, pro, marker=markers[i], linestyle='-', color=colors[i], linewidth=1.2, label=labels[i])

# 
axs[0].set_xlabel('Time', **label_font_conf)
axs[0].set_ylabel('Percentage (%)', **label_font_conf)
axs[1].set_xlabel('Time', **label_font_conf)
axs[1].set_yticks([])

# 
axs[0].grid(axis='y', alpha=0.3)
axs[1].grid(axis='y', alpha=0.3)

axs[0].tick_params(labelsize=15)
axs[0].tick_params(axis='x', rotation=45)
axs[1].tick_params(labelsize=15)
axs[1].tick_params(axis='x', rotation=45)

# 
plt.tight_layout()
axs[0].legend(loc='lower left', ncol=3, bbox_to_anchor=(0, 1), frameon=False, columnspacing=4,fontsize=16)
# plt.show()
plt.savefig('./figure4.pdf', bbox_inches="tight")
```
